## Evaluate the Base Model

```python
## Create the questions
questions = [
    "How can I cancel my order?",
    "How do I track my shipment?",
    "When will I receive my refund?",
    "How do I replace a damaged product?",
    "Why did my payment fail?",
    "Can I change my delivery address?",
    "How long does shipping take?",
    "Can I return a product after delivery?",
    "How do I contact customer support?",
    "What should I do if I receive the wrong item?"
]
```

```python
## Create a generation function (Lama3.2B)
def generate_answer(question):

    messages = [
        {
            "role": "user",
            "content": question,
        }
    ]

    inputs = tokenizer.apply_chat_template(
        messages,
        tokenize=True,
        add_generation_prompt=True,
        return_tensors="pt",
    ).to("cuda")

    outputs = model.generate(
        inputs,
        max_new_tokens=150,
        temperature=0.2,
        do_sample=False,
    )

    response = tokenizer.decode(outputs[0], skip_special_tokens=True)

    return response
```

```python
## Generate answers
answers = []

for q in questions:

    print("=" * 80)
    print("QUESTION:", q)

    ans = generate_answer(q)

    answers.append(ans)

    print(ans)
    print()
```

```python
from datasets import load_dataset

dataset = load_dataset(
    "json",
    data_files="/kaggle/input/notebooks/marree/non-instruction-finetuning-ipynb/instruction_dataset.jsonl",
    split="train",
)

print(dataset)
```

```python
## Format the Dataset
EOS_TOKEN = tokenizer.eos_token

def formatting_prompts_func(examples):
    instructions = examples["instruction"]
    responses = examples["response"]

    texts = []

    for instruction, response in zip(instructions, responses):

        text = f"""### Instruction:
{instruction}

### Response:
{response}{EOS_TOKEN}"""

        texts.append(text)

    return {"text": texts}

dataset = dataset.map(
    formatting_prompts_func,
    batched=True,
)

print(dataset)
```

## Apply LoRA

```python
model = FastLanguageModel.get_peft_model(
    model,
    r=16,
    target_modules=[
        "q_proj",
        "k_proj",
        "v_proj",
        "o_proj",
        "gate_proj",
        "up_proj",
        "down_proj",
    ],
    lora_alpha=16,
    lora_dropout=0,
    bias="none",
    use_gradient_checkpointing="unsloth",
    random_state=3407,
)

print("✅ LoRA applied successfully")
```

## Create the SFT Trainer

```python
## Unsloth 2026.6.9,TRL 0.24.0 ,Transformers 5.5.0,Tesla T4

from trl import SFTTrainer, SFTConfig
```

## Configure training

```python
trainer = SFTTrainer(
    model=model,
    tokenizer=tokenizer,
    train_dataset=dataset,

    dataset_text_field="text",

    args=SFTConfig(
        output_dir="/kaggle/working/sft_outputs",

        max_seq_length=2048,

        per_device_train_batch_size=2,
        gradient_accumulation_steps=4,

        learning_rate=2e-4,

        num_train_epochs=1,

        logging_steps=20,

        save_strategy="epoch",

        optim="adamw_8bit",

        lr_scheduler_type="cosine",

        warmup_ratio=0.03,

        report_to="none",

        seed=3407,
    ),
)
```

## Training

```python
trainer.train()
```

```python
## Create a cleaned dataset placeholder

import json

replacements = {
    "{{Order Number}}": "ORDER_NUMBER",
    "{{Customer Support Hours}}": "customer support hours",
    "{{Customer Support Phone Number}}": "customer support phone number",
    "{{Online Company Portal Info}}": "customer account portal",
    "{{Online Order Interaction}}": "Orders section",
}

input_file = "/kaggle/input/notebooks/marree/non-instruction-finetuning-ipynb/instruction_dataset.jsonl"
output_file = "/kaggle/working/instruction_dataset_clean.jsonl"

count = 0

with open(input_file, "r", encoding="utf-8") as fin, \
     open(output_file, "w", encoding="utf-8") as fout:

    for line in fin:
        item = json.loads(line)

        instruction = item["instruction"]
        response = item["response"]

        for old, new in replacements.items():
            instruction = instruction.replace(old, new)
            response = response.replace(old, new)

        json.dump(
            {
                "instruction": instruction,
                "response": response,
            },
            fout,
            ensure_ascii=False,
        )
        fout.write("\n")

        count += 1

print(f"Cleaned {count} examples")


```

## Generate preference_dataset.jsonl

```python
import json
import random

bad_responses = [
    "Please contact customer support.",
    "I don't know.",
    "Try again later.",
    "The order cannot be cancelled.",
    "Refunds are not available.",
    "No information is available.",
    "Contact your bank.",
    "Please read the website.",
    "Your request cannot be processed.",
    "We cannot help with that."
]

input_file = "/kaggle/working/instruction_dataset_clean.jsonl"
output_file = "/kaggle/working/preference_dataset.jsonl"

count = 0

with open(input_file, "r", encoding="utf-8") as fin, \
     open(output_file, "w", encoding="utf-8") as fout:

    for line in fin:
        item = json.loads(line)

        example = {
            "prompt": item["instruction"],
            "chosen": item["response"],
            "rejected": random.choice(bad_responses)
        }

        fout.write(json.dumps(example, ensure_ascii=False) + "\n")
        count += 1

print(f"Created {count} preference examples")
```
